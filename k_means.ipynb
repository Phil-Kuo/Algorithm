```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import sklearn.datasets as ds
```

```python
# 载入iris数据集
dataset = ds.load_iris();
fig = plt.figure(figsize=(6, 6), dpi=80)
ax = fig.add_subplot(1, 1, 1)
ax.scatter(dataset.data[:,1], dataset.data[:,2])
```

```python
def visualize(ax, data, labels, centers):
    """
    将模型结果可视化
    """
    colors = ["#82CCFC", "k"]
    if labels is None:
        ax.scatter(data[:, 0], data[:, 1])
    else:
        ax.scatter(data[:, 0], data[:, 1], c=[colors[i] for i in labels],
            marker="o", alpha=0.6)
    ax.scatter(centers[:, 0], centers[:, 1], marker="*", c=colors, edgecolors="white",
        s=700., linewidths=2)
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
```

```python
data = dataset.data[:,1:3]
fig = plt.figure(figsize=(10, 10), dpi=80)
for i in range(4):
    ax = fig.add_subplot(2, 2, i+1)
    if i == 0:
        visualize(ax, data, None, np.array([[2, 5], [3, 1]]))
    else:
        model = KMeans(n_clusters=2, max_iter=i, init='k-means++',n_init=1).fit(data)
        visualize(ax, data, model.labels_, model.cluster_centers_)
plt.show()
```
